```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Excel file readen
df = pd.read_excel('Data Overzicht.xlsx', sheet_name=3)
Nagalmtijd = df['Gemiddelde Nagalmtijd']
Positie_x = df['Stoel']
Positie_y = df['Rij']

# Dataset
x = np.array(Positie_x)
y = np.array(Positie_y)
z = np.array(Nagalmtijd)
colo = Nagalmtijd

# Grafiek
fig = plt.figure(figsize=(13, 13))
ax = fig.add_subplot(111, projection='3d')

# Kleurschaal
color_map = cm.ScalarMappable(cmap=cm.coolwarm)
color_map.set_array(colo)


# Heatmap zelf
ax.plot_trisurf(x, y, z, cmap=cm.coolwarm, alpha=0.8)

# Assen limieten
ax.set_xlim(1, 8)
ax.set_ylim(7, 1)
ax.set_zlim(0, 1.5)

# Titel en assen
ax.set_title("Pathe City zaal 1", fontsize=30)
ax.set_xlabel('Stoelnummer', fontsize=17)
ax.set_ylabel('Rij', fontsize=17)
ax.set_zlabel('Nagalmtijd(s)', fontsize=12)

# De display
plt.show()

```

```python
# Data omzetten in een pandas frame
data = pd.DataFrame({
    "y":Positie_y,
    "x":Positie_x,
    "z":Nagalmtijd
}).pivot_table("z", "y", "x")

#Heatmap maken
fig, ax = plt.subplots()
sns.heatmap(data, vmin=0, cmap=cm.jet, cbar_kws={"label": "Nagalmtijd (s)"})


#Heatmap titels
ax.set_title("Pathe City zaal 1")
ax.set_xlabel('Stoelnummer')
ax.set_ylabel('Rij')

#Show de Heatmap
plt.show()
```

```python
# Data omzetten in een pandas frame
Std= df['Standaarddeviatie']
Standaarddeviatie = pd.DataFrame({
    'y':Positie_y,
    'x':Positie_x,
    'z':Std
}).pivot_table("z", "y", "x")

#Heatmap maken
fig, ax = plt.subplots()
sns.heatmap(Standaarddeviatie, cmap=cm.jet, cbar_kws={"label": "Standaarddeviatie"})


#Heatmap titels
ax.set_title("Pathe City zaal 1")
ax.set_xlabel('Stoelnummer')
ax.set_ylabel('Rij')

#Show de Heatmap
plt.show()
```
